```python
import numpy as np
```

```python
# Step 1: of 120 values using seed number 5
#step2: reshape the array to a 2D array with 12 rows and 10 columns
# step 3: select the values from rows 5 to 8 and columns 4 to 6
#step 4: broadcast the value 32 all over the selected values
np.random.seed(5)
arr=np.random.randint(0,100,120)
arr_2D= arr.reshape(12,10)

selected_values= arr_2D[5:9,4:7]
selected_values[:]=32

#step 5: Calculate the mean,meadian,25,75 and 85th percentile
mean_value=np.mean(selected_values)
median_value=np.median(selected_values)
Average_value=np.average(selected_values)
percentile=np.percentile(selected_values,[25,75,85])
```
