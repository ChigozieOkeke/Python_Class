```python
# What is the average crude oil price over the entire dataset?
average=df["Crude Oil Price"].mean()
print("The Crude Oil Price Average is:", average)
```

```python
# What is the maximum production (in million barrels per day) recorded in the dataset, and in which month and year did it occur?
max_production=df["Production"].max()
print("The maximum production is:", max_production)
```

```python
# What is the maximum production (in million barrels per day) recorded in the dataset, and in which month and year did it occur?
df.apply('max')
```

```python
# What is the minimum export volume (in million barrels per day) recorded in the dataset, and in which month and year did it occur?
min_export_volume=df["Crude Oil Export"].min()
print("The minimum export volume is:", min_export_volume)
```

```python
# What is the minimum export volume (in million barrels per day) recorded in the dataset, and in which month and year did it occur?
df.apply('min')
```

```python
# How many months had a crude oil price above $100 per barrel
months_above_100=df[df["Crude Oil Price"]>100].shape[0]
print("Number of months with crude oil price above $100 :", months_above_100)
```

```python
# What is the correlation between crude oil price and production volume?
correlation=df["Crude Oil Price"].corr(df["Production"])
print("The correlation between Crude oil price and production is:", correlation)
```

```python
# What is the average export volume (in million barrels per day) for months with a crude oil price below $50 per barrel?
average=df["Crude Oil Export"].mean()
print("The Crude Oil Export  Average is:", average)
```

```python
# What is the average export volume (in million barrels per day) for months with a crude oil price below $50 per barrel?
df.apply("mean")
```
