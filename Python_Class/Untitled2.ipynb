```python
import pandas as pd
import numpy as np
```

```python
## writting the summation of a paticular column at the bottom of the row
data={"A":[1,2,3,4,5], ## data contains both the columns and its row value
      "B":[10,20,30,40,50]}
df=pd.DataFrame(data)
sum_total=df["B"].sum()
df.loc["Total"]=pd.Series(df["B"].sum(),index=["B"])
df
```

```python
data={"A":[1,2,3,4,5], 
      "B":[10,20,30,40,50],
      "C":[100,200,300,400,500]}
df=pd.DataFrame(data)
for col in df.columns:
      total=df[col].sum()
      df.at[df.index[-1],col]=total
      print(df)
```

```python
data={"Names":["Emeka","Chika","John","Goodnes","Sharon"], 
      "English Score":[10,20,30,40,50],
      "Maths Score":[100,200,300,400,500],
      "Biology Score":[102,28,300,40,50]}
df=pd.DataFrame(data)
sum_eng=df["English Score"].sum()
sum_mat=df["Maths Score"].sum()
sum_bio=df["Biology Score"].sum()
df.loc[len(df)]=["Total",sum_eng,sum_mat,sum_bio]## adding Total at the bottom of the Names Column
df
```

```python
data={"Names":["Emeka","Chika","John","Goodnes","Sharon"], 
      "English Score":[10,20,30,40,50],
      "Maths Score":[100,200,300,400,500],
      "Biology Score":[102,28,300,40,50]}
df=pd.DataFrame(data)
sum_eng=df["English Score"].mean()
sum_mat=df["Maths Score"].mean()
sum_bio=df["Biology Score"].mean()
df.loc[len(df)]=["Average",sum_eng,sum_mat,sum_bio]## adding Total at the bottom of the Names Column
df
```
