```python
#A brief introduction to machine learning and its types.
#---

# Machine learning is a subset of artificial intelligence (AI) that enables computer systems 
# to automatically learn and improve from experience without being explicitly programmed. 
# It focuses on the development of algorithms and statistical models that enable machines 
# to perform tasks without explicit instructions.

# There are three main types of machine learning:

# 1. **Supervised Learning**: 
# In this type of learning, the model is trained on a labeled dataset, 
# where the input data and corresponding output labels are provided. 
# The goal is to learn a mapping from inputs to outputs and make predictions on new, unseen data.

# 2. **Unsupervised Learning**: In unsupervised learning, the model is trained on an unlabeled dataset, 
# where only the input data is provided. The goal is to identify patterns, relationships, or structures
# within the data.

# 3. **Reinforcement Learning**: In reinforcement learning, an agent learns to make decisions by
# interacting with an environment. The agent receives rewards or penalties based on its actions, 
# and the goal is to learn a policy that maximizes the cumulative reward over time.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
## lets download data set from kaggle.com
```

```python
df=pd.read_csv("File_OLX_cars_dataset00.csv")
df.head()
```

```python
df
```

## DATA PROCESSING

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(["Car Profile"], axis=1, inplace=True)
df.isnull().sum()
```

```python
df.drop(["Images URL's"],axis=1,inplace=True)
df.isnull().sum()
```

```python
df.columns
```

```python
df.head()
```

```python
df.drop(["Description"],axis=1,inplace=True)
df.isnull().sum()
```

```python
df.drop(["Seller Location"],axis=1,inplace=True)
df.isnull().sum()
```

```python
df.drop(["Ad ID"],axis=1,inplace=True)
df.isnull().sum()
```

```python
df.columns
```

## REMOVING PUNTUATION MARKS OR SYMBOLS

```python
df["Car Features"]= df["Car Features"].str.replace(",","")
df.info()
## it means ",","" "" means replace those coma with empty space 
```

```python
df.head()
```

```python
## labelencoder convert categorical values to numeric thats why we import it
## convert categorical values to numerical values
encoder=LabelEncoder()
category=df.select_dtypes(include=["object","category"])
for x in category:
    df[x]=encoder.fit_transform(df[x])
df.head()
```

## Feature Engineering

```python
## split the dataset into x and y 
##  thats dependent and independent variable, the X should be capital letter 
## X is independent variables 
X=df.drop(["Price"],axis=1) #independent variables
y=df["Price"] # dependent variables
```

```python
X.head()
```

```python
y.head()
```

```python
## check for feature thats to know which one is important
from sklearn.feature_selection import mutual_info_classif
plt.style.use("ggplot")
importance=mutual_info_classif(X,y)
feature_impo=pd.Series(importance,X.columns[0:len(X.columns)])
feature_impo.plot(kind="bar",color="purple")
plt.show()
```

```python
## check for feature thats to know which one is important
## change to horizontal by adding h to the bar
from sklearn.feature_selection import mutual_info_classif
plt.style.use("ggplot")
importance=mutual_info_classif(X,y)
feature_impo=pd.Series(importance,X.columns[0:len(X.columns)])
feature_impo.plot(kind="barh",color="purple")
plt.show()
```

```python
X=df.drop(["Condition"],axis=1)
```

```python
X.head()
```

```python
from sklearn.feature_selection import mutual_info_classif
plt.style.use("ggplot")
importance=mutual_info_classif(X,y)
feature_impo=pd.Series(importance,X.columns[0:len(X.columns)])
feature_impo.plot(kind="barh",color="purple")
plt.show()
```

```python
## using model selection into train test split
## spliting the data set into training and testing set 
Xtrain, xtest,ytrain,ytest=train_test_split(X,y,test_size=0.20,random_state=1)
## 0.20 is 20 percent, 1.20 means 100 percent
## 20 percent for testing and 80 percent for training, training percent has to be bigger than testing percent
```

## NOW FIT IN THE DATA SET

```python
from sklearn.pipeline import Pipeline
LinearModel=Pipeline([("Scaler",StandardScaler()),("Regression",LinearRegression())])
print("LinearRegression in process.....")
LinearModel.fit(Xtrain,ytrain)
print("Done fitting Training set.")
```

```python
pred1=LinearModel.predict(xtest)
pred1
```

```python
## Checking for the accuracy of of the Machine Learning
print("MAE",mean_absolute_error(ytest,pred1))
print("MAPE",mean_absolute_percentage_error(ytest,pred1))
print("MSE",mean_squared_error(ytest,pred1))
print("R2 SCORE",r2_score(ytest,pred1))
```

```python
outcome={
    "Actual Price":ytest,
    "predicted Price":pred1
}
df3=pd.DataFrame(outcome)
df3.head()
```

```python
sns.regplot(x=ytest,y=pred1)
```

```python
df3.to_csv("outcome")
```
