```python
# The code `df.drop("NILL", axis=1, inplace=True)` is used to drop a column named "NILL" from a pandas DataFrame `df`. 

#Here is an explanation of the parameters used in this code:
#- `df`: This is the pandas DataFrame from which we want to drop the column.
#- `"NILL"`: This is the label of the column that we want to drop.
#- `axis=1`: This indicates that we are referring to columns. In pandas DataFrame, `axis=1` refers to columns and `axis=0` refers to rows.
#- `inplace=True`: This parameter makes sure that the operation is done on the DataFrame itself rather than returning a new DataFrame. If `inplace=True`, the original DataFrame `df` will be modified. If `inplace=False` (or it is not specified), then a new DataFrame will be returned without modifying the original DataFrame.

#After executing this code, the column with the label "NILL" will be dropped from the DataFrame `df`.
df.drop("NILL",axis=1,inplace=True)
df
```

```python
# The code `plt.plot(df["Year"], df["CrudeOil"])` is used to create a line plot using the matplotlib library in Python.
# Here's a breakdown of each part of the code:
#-`plt`: This refers to the `pyplot`module from the `matplotlib` library, which is commonly used for creating 
# visualizations  like plots.
# - `plot()`: This is a function from `pyplot` that is used to create a line plot.
# - `df["Year"]`: This is accessing the "Year" column from a DataFrame `df`, assuming that `df` is a pandas DataFrame.
# - `df["CrudeOil"]`: This is accessing the "CrudeOil" column from the same DataFrame.

# Putting it together, `plt.plot(df["Year"], df["CrudeOil"])` will create a line plot where the x-axis corresponds to the
# values in the "Year" column and the y-axis corresponds to the values in the "CrudeOil" column from the DataFrame `df`.
plt.plot(df["Year"],df["CrudeOil"])
```

```python
#When you call `df["Period"].unique()`, it will give you an array containing all the unique values
# in the "Period" column of the DataFrame `df`. 
# This method is useful when you want to see all the distinct values present in a specific column of your DataFrame.
df["Period"].unique() # it use to see the different unique values in a column
```

```python
# This code snippet is filtering a DataFrame `df` to select rows where the "Period" column has the value of "Annual". 

# Here's a breakdown of the code:
#- `df["Period"]`: This selects the "Period" column from the DataFrame `df`.
# - `df["Period"]=="Annual"`: This condition checks for each row in the "Period" column whether the value is equal to "Annual". This results in a boolean Series with `True` for rows where the condition is met and `False` otherwise.
#- `df[...]`: This indexing operation takes the boolean Series resulting from the condition and selects only the rows of `df` where the condition is `True`, effectively filtering out rows where the condition is `False`.

# So, the final result would be a DataFrame containing only the rows where the period is "Annual".
# this is use in Column where there is diferent value and this is need to call out specific value out of the rest
df[df["Period"]=="Annual"]
df.head(3)# we filtered only Anual from the rest of the values in the column "Period" in table 
```

```python
#This code snippet is selecting rows from a DataFrame `df`
# where the value in the "Period" column is equal to "Annual", and assigning the result to a new DataFrame `df2`.
# Finally, it displays the first 3 rows of the new DataFrame `df2`.

# So, `df2 = df[df["Period"]=="Annual"]` filters out only the rows where the "Period" column has the value "Annual", and `df2.head(3)` then shows the first 3 rows of the resulting DataFrame `df2`.
# the filtered Table is Assigned to a new data frame
df2=df[df["Period"]=="Annual"]
df2.head(3) ## Below Result is the filtered Table and the column Period has only the filtered value "Anual"
```

```python
# The code `plt.plot(df2["Year"], df2["CrudeOil"])` is used to create a line plot using the `matplotlib` library in Python.
# Here's a breakdown of what this code does:

#- `plt.plot()`: This function is used to create a line plot. It takes two main arguments:
# 1 - `df2["Year"]`: This specifies the data to be plotted on the x-axis. 
# In this case, it seems like the "Year" column of the DataFrame `df2` is used for the x-axis values.
# 2 - `df2["CrudeOil"]`: This specifies the data to be plotted on the y-axis. 
# It seems like the "CrudeOil" column of the DataFrame `df2` is used for the y-axis values.

# Therefore, `plt.plot(df2["Year"], df2["CrudeOil"])` creates a line plot where the x-axis values are taken from the "Year" column of `df2` and the y-axis values are taken from the "CrudeOil" column of `df2`. Each point on the plot will correspond to one year from the "Year" column and the value of crude oil for that year from the "CrudeOil" column.
plt.plot(df2["Year"],df2["CrudeOil"])## ploting the Year against Crude in the new filtered table
```
