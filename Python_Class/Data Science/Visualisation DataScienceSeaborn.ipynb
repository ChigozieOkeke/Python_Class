```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("File_tips.csv")
df
```

```python
plt.style.use("ggplot")
sns.displot(x=df["Total_bill"],kde=True)
#This code sets the style of the plot to "ggplot" 
# and creates a histogram of the "Total_bill" column
# of the dataframe df with a kernel density estimate (kde) curve.
# The histogram shows the distribution of the total bill amounts 
# in the dataset.
# The kde curve is a smoothed estimate of the probability density function,
# which can help identify the underlying distribution of the data.
```

```python
sns.regplot(x=df["Total_bill"],y=df["Tip"])
```

```python
sns.regplot(x=df["Total_bill"],y=df["Size"])
```

```python
sns.scatterplot(x=df["Total_bill"],y=df["Tip"],palette='colorblind',hue=df["Day"])
```

```python
sns.scatterplot(x=df["Total_bill"],y=df["Size"])
```

```python
sns.countplot(x=df["Day"],color="blue")
```

```python
plt.figure(figsize=(10,5))
ln=sns.barplot(x="Sex",y="Total_bill",data=df,palette="Purples",hue="Day")
```

```python
## 
plt.figure(figsize=(10,5))
ln=sns.barplot(x="Sex",y="Total_bill",data=df,palette="Greens",estimator="max",hue="Day")

for p in ln.patches:
    ln.annotate(f"{p.get_height()}",(p.get_x()  + p.get_width()/2.,p.get_height()),
                ha="center",va="baseline",fontsize=10,color="black",xytext=(0,5),
                textcoords="offset points")
    
plt.title("Day Distribution")
plt.xticks(rotation=45)
plt.legend(loc=[1,0])
plt.show()
```
