```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("sample_pivot.csv")
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
## forward fill
df1=df.copy()
df1.fillna(method="ffill",inplace=True)
df1
```

```python
## backward fill
df1=df.copy()
df1.fillna(method="bfill",inplace=True)
df1
```

```python
df3=df.copy()
df3.fillna(np.mean(df["Units"]),inplace=True)
df3
```

```python
df.dropna(inplace=True)
```

```python
## pivot table
df_pv=pd.pivot_table(df,index="Region",columns="Type",values=["Units","Sales"],aggfunc="count")
df_pv
```

```python
df_pv=pd.pivot_table(df,index="Region",columns="Type",values=["Units","Sales"],aggfunc="max",margins=True)
df_pv
```

```python
df.sort_values("Date",ascending=True,ignore_index=True,inplace=True)
df
```

```python
east=df[df["Region"]=="East"]
west=df[df["Region"]=="West"]
north=df[df["Region"]=="North"]
south=df[df["Region"]=="South"]
```

```python
east
```

```python
south
```

```python
alldf=pd.concat([east,west,north,south]) ##we add different data frame together
alldf
```

```python
alldf.to_csv("Region Analysis.csv",index=False)
```

```python
alldf.to_excel("Region Analysis.xlsx",index=False,sheet_name="Sales")
```

```python
## saving each excel tables into different sheet.
with pd.ExcelWriter("AllSales.xlsx",engine="xlsxwriter") as writer:
    east.to_excel(writer,sheet_name="East",index=False,engine="xlsxwriter")
    west.to_excel(writer,sheet_name="West",index=False,engine="xlsxwriter")
    north.to_excel(writer,sheet_name="North",index=False,engine="xlsxwriter")
    south.to_excel(writer,sheet_name="South",index=False,engine="xlsxwriter")
    
```
